```python
import pandas as pd
import numpy as np
df = pd.read_csv("customer_segmentation_data.csv")

df.info()
```

```python
df = df.drop(columns=["id"])
df.head()
```

```python
df[["age","income","spending_score","membership_years","purchase_frequency","last_purchase_amount"]] = df[["age","income","spending_score","membership_years","purchase_frequency","last_purchase_amount"]].replace(0,np.nan)
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ['age', 'income', 'spending_score', 'membership_years', 'purchase_frequency', 'last_purchase_amount']
df_scaled = df.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df_scaled.head()
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

# --- 1. Histograms for numerical columns ---
numeric_cols = ['age', 'income', 'spending_score', 'membership_years', 
                'purchase_frequency', 'last_purchase_amount']

df[numeric_cols].hist(bins=20, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
plt.savefig("numerical_features_distributions.png")
```

```python
bins = [17, 30, 50, 70]   
labels = ['Young (18-30)', 'Middle-aged (31-50)', 'Older (51-70)']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

age_counts = df['age_group'].value_counts(normalize=True) * 100
#print(age_counts)

age_counts.plot(kind='bar', color='steelblue')
plt.title("Customer Distribution by Age Group")
plt.ylabel("Percentage of Customers")
plt.show()

```

```python
from sklearn.cluster import KMeans
import numpy as np

X = df_scaled[['income', 'spending_score']]  # scaled features

# Elbow Method
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method For Optimal k")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X)  # add cluster labels to original df
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['income'], df['spending_score'], 
            c=df['cluster'], cmap='viridis', alpha=0.6, edgecolor='k')
plt.title("Customer Segments (Income vs Spending Score)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# Create summary table with raw (unscaled) values
cluster_summary = df.groupby("cluster").agg({
    "age": "mean",
    "income": "mean",
    "spending_score": "mean",
    "membership_years": "mean",
    "purchase_frequency": "mean",
    "last_purchase_amount": "mean",
    "preferred_category": lambda x: x.value_counts().idxmax()  # most common category
}).reset_index()

# Rename columns for readability
cluster_summary = cluster_summary.rename(columns={
    "age": "Avg Age",
    "income": "Avg Income",
    "spending_score": "Avg Spending Score",
    "membership_years": "Avg Membership Years",
    "purchase_frequency": "Avg Purchase Frequency",
    "last_purchase_amount": "Avg Last Purchase",
    "preferred_category": "Top Category"
})

# Map numeric cluster IDs to business-friendly names
cluster_labels = {
    0: "Value Enthusiasts (Low Income, High Spending)",
    1: "Untapped Potential (High Income, Low Spending)",
    2: "Budget Occasionals (Low Income, Low Spending)",
    3: "Premium Loyalists (High Income, High Spending)",
    4: "Core Steady (Mid Income, Mid Spending)"
}
cluster_summary["Cluster Name"] = cluster_summary["cluster"].map(cluster_labels)

# Drop the numeric cluster column
cluster_summary = cluster_summary.drop(columns=["cluster"])

# Reorder columns
cluster_summary = cluster_summary[
    ["Cluster Name", "Avg Age", "Avg Income", "Avg Spending Score", 
     "Avg Membership Years", "Avg Purchase Frequency", "Avg Last Purchase", "Top Category"]
]

# Nice formatting for floats
import pandas as pd
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")

# Show the table
display(cluster_summary)

# Export to Markdown for README
# print(cluster_summary.to_markdown(index=False))
```
